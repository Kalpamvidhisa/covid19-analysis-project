```python
# ================================
# 📌 COVID-19 DATA ANALYSIS PROJECT
# ================================

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 📥 Step 1: Load Dataset
# -------------------------------
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"

df = pd.read_csv(url)

# -------------------------------
# 🌍 Step 2: Prepare Country-wise Data
# -------------------------------
date_cols = df.columns[4:]

df_country = df.groupby("Country/Region")[date_cols].sum()
df_country = df_country.T

df_country.index = pd.to_datetime(df_country.index)

# -------------------------------
# 📊 Step 3: Daily & Weekly Cases
# -------------------------------
daily_cases = df_country.diff().fillna(0)
weekly_cases = daily_cases.rolling(7).sum()

# -------------------------------
# 📉 Step 4: Rolling Average
# -------------------------------
rolling_avg = daily_cases.rolling(7).mean()

# -------------------------------
# 🌍 Step 5: Country Comparison Plot
# -------------------------------
countries = ["India", "US", "Brazil"]

plt.figure()
rolling_avg[countries].plot()
plt.title("7-Day Rolling Average COVID Cases")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.show()

# -------------------------------
# 🔺 Step 6: Peak Detection
# -------------------------------
print("\nPeak Dates:")
for country in countries:
    peak_date = rolling_avg[country].idxmax()
    print(f"{country}: {peak_date}")

# -------------------------------
# 🧠 Step 7: Reproduction Rate (Rt)
# -------------------------------
Rt = daily_cases / daily_cases.shift(7)

plt.figure()
Rt[countries].plot()
plt.axhline(y=1, linestyle='--')
plt.title("Approximate Reproduction Rate (Rt)")
plt.xlabel("Date")
plt.ylabel("Rt Value")
plt.show()

# -------------------------------
# 📤 Step 8: Save Outputs
# -------------------------------
rolling_avg[countries].plot()
plt.savefig("covid_trend.png")

Rt[countries].plot()
plt.savefig("covid_rt.png")

print("\n✅ Charts saved successfully!")
```
